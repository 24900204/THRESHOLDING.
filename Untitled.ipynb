```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Piano_Sheet_Music.png', cv2.IMREAD_GRAYSCALE)
```

```python
retval, img_thresh_gbl_1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
retval, img_thresh_gbl_2 = cv2.threshold(img, 130, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=[18, 15])
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
```

```python
plt.subplot(222)
plt.imshow(img_thresh_gbl_1, cmap='gray')
plt.title('Global Threshold 50')
plt.axis('off')
```

```python
plt.subplot(223)
plt.imshow(img_thresh_gbl_2, cmap='gray')
plt.title('Global Threshold 130')
plt.axis('off')
```

```python

if len(img.shape) == 3:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
else:
    gray = img

# Adaptive Threshold
img_thresh_adp = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

```

```python
plt.subplot(224)
plt.imshow(img_thresh_adp, cmap='gray')
plt.title('Adaptive Threshold')
plt.axis('off')
```
